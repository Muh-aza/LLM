```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the genes for each source
genes_data = {
    "True": ["CDH1", "CTNNB1", "CTNNA", "CTNND1", "CDC42", "CDH3", "ACTN4", "ACTB", "EP300", "FGFR1", "LEF1", "PARD3", "MET", "LEF1", "SMAD4"],
    "Claude_": ["CDH1", "CTNNB1", "CTNNA", "CTDNND1", "CDC", "CDH3", "ACTN1", "ACT", "PVRL1", "MLLT4", "LIMA1", "PRKCA", "MLLT4", "VCL", "DSP"],
    "Claude instant Ans": ["CDH1", "CTNNB1", "CTNNA", "CTNND1", "CDC", "CDH3", "ACTN", "ACT", "PLEKHA7", "RHOA", "LMO7", "DSG1", "RAC1", "VCL", "AFADIN"],
    "Cohere": ["CDH1", "CTNNB1", "CTNNA1", "CTNND1", "CDC42", "CDH3", "ACTN4", "ACTB", "EPHA2", "FBLN1", "LED1", "PARD3", "MYC", "PLXNA2", "SND1"],
    "Chat-Gpt-3.5": ["CDH1", "CTNNB1", "CTNNA1", "CTNND1", "Cdc42", "JUP", "Rac1", "ZO-2", "JAM", "ARVCF", "ZO-2", "Zo1", "RHOA", "VCL", "AFADIN"]
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges for "Claude_" source
source = "Claude_"
for gene in genes_data[source]:
    G.add_node(gene)
    G.add_edge(source, gene)

# Create a layout for the nodes
pos = nx.spring_layout(G, seed=42)

# Draw the graph for "Claude_"
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=800, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray', arrows=True)
plt.title(f"Genes Involved in Adherens Junction for {source}")
plt.show()

# Create a directed graph for the remaining values
remaining_values = {src: genes for src, genes in genes_data.items() if src != source}

# Create a new directed graph for the remaining values
G_remaining = nx.DiGraph()

# Add nodes and edges for the remaining values
for src, genes in remaining_values.items():
    for gene in genes:
        G_remaining.add_node(gene)
        G_remaining.add_edge(src, gene)

# Create a layout for the nodes of the remaining values graph
pos_remaining = nx.spring_layout(G_remaining, seed=42)

# Draw the graph for the remaining values
plt.figure(figsize=(12, 8))
nx.draw(G_remaining, pos_remaining, with_labels=True, node_size=800, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray', arrows=True)
plt.title("Genes Involved in Adherens Junction for Remaining Sources")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['Claude-1 (prompt_1)', 'Claude-1 (prompt_2)', 'Claude-instant', 'Chat-GPT3.5', 'Cohere']
accuracy = [0.3200, 0.3550, 0.3200, 0.0850, 0.4000]
precision = [0.9667, 0.8914, 0.8993, 0.9000, 0.5802]
recall = [0.3200, 0.3550, 0.3200, 0.0850, 0.4000]
f1_score = [0.4014, 0.4647, 0.3962, 0.1467, 0.4698]

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Performance Metrics', fontsize=16)

# Bar width
bar_width = 0.2
index = np.arange(len(models))

# Function to add values on top of bars
def add_values(ax, data):
    for i, v in enumerate(data):
        ax.text(i, v + 0.01, f'{v:.4f}', ha='center', va='bottom', fontsize=10)

# Subplot 1: Accuracy
axs[0, 0].bar(index, accuracy, bar_width, label='Accuracy', color='skyblue')
axs[0, 0].set_title('Accuracy')
axs[0, 0].set_xticks(index)
axs[0, 0].set_xticklabels(models, rotation=45, ha='right')
axs[0, 0].set_ylim(0, 1)
add_values(axs[0, 0], accuracy)

# Subplot 2: Precision
axs[0, 1].bar(index, precision, bar_width, label='Precision', color='lightcoral')
axs[0, 1].set_title('Precision')
axs[0, 1].set_xticks(index)
axs[0, 1].set_xticklabels(models, rotation=45, ha='right')
axs[0, 1].set_ylim(0, 1)
add_values(axs[0, 1], precision)

# Subplot 3: Recall
axs[1, 0].bar(index, recall, bar_width, label='Recall', color='lightgreen')
axs[1, 0].set_title('Recall')
axs[1, 0].set_xticks(index)
axs[1, 0].set_xticklabels(models, rotation=45, ha='right')
axs[1, 0].set_ylim(0, 1)
add_values(axs[1, 0], recall)

# Subplot 4: F1 Score
axs[1, 1].bar(index, f1_score, bar_width, label='F1 Score', color='lightsalmon')
axs[1, 1].set_title('F1 Score')
axs[1, 1].set_xticks(index)
axs[1, 1].set_xticklabels(models, rotation=45, ha='right')
axs[1, 1].set_ylim(0, 1)
add_values(axs[1, 1], f1_score)

# Adjust layout
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Add legends
for ax in axs.flat:
    ax.legend()

# Show the plot
plt.show()
```
